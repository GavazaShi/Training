Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

Data processing

```python
XColumns= ["weight", "waist", "pulse"]
yColumns = ["chins", "situps", "jumps"]
columnNumber = 0
targetColumnNumber = 1
y,X = datasets.load_linnerud(return_X_y= True)
print(X.shape)
print(y.shape)
print(X)
print(y)
```

```python

X= X[:,columnNumber]
print(X.shape)
X = X.reshape((-1,1))
print(X.shape)

y = y[:,targetColumnNumber]
print(y.shape)
y= y.reshape((-1,1))
print(y.shape)
```

Splitting training and test data and training the model

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.75)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Predicting, Plotting the data and determining accuracy

```python
y_pred = model.predict(X_test)
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, y_pred, color="green", linewidth=3)
plt.xlabel(f'{XColumns[columnNumber]}')
plt.ylabel(f'{yColumns[targetColumnNumber]}')
plt.title(f'A Graph Plot Showing Prediction of {yColumns[targetColumnNumber]} by {XColumns[columnNumber]}')
plt.plot()

```
